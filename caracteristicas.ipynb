```python
def histograma2(img, bins):
    red, b = np.histogram(img[:,:,0], bins)
    green, b = np.histogram(img[:,:,1], bins)
    blue, b  = np.histogram(img[:,:,2], bins)
    h = np.array([red, green, blue])
    return h
```

```python
h = histograma2(img)
```

```python
h2 = h
```
