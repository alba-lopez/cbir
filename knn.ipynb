```python
def histograma_lab(image, bins, mask=None):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)  # Convierte la imagen a espacio L*a*b*.
    histogram = cv2.calcHist([lab], [0, 1, 2], mask, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    histogram = cv2.normalize(histogram, histogram).flatten()
    return histogram
```

```python
def histograma_rgb(image, bins, mask=None):
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convierte la imagen a espacio L*a*b*.
    histogram = cv2.calcHist([lab], [0, 1, 2], mask, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    histogram = cv2.normalize(histogram, histogram).flatten()
    return histogram
```
