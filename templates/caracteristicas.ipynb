```python
import numpy as np
import cv2
```

```python
filename = ".\static\pug.jpeg"
```

```python
img = cv2.imread(filename, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def histograma(img):
    h, b = np.histogram(img, 256)
    return h
```

```python
h = histograma(img)
```

```python
def histograma2(img):
    red = np.histogram(img[:,:,0], 255)
    green = np.histogram(img[:,:,1], 255)
    blue = np.histogram(img[:,:,2], 255)
    return (red, green, blue)
```

```python
r, g, b = histograma2(img)
```
